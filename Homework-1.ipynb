```python
import sys
import os

print(sys.version)

############################################################
# Int.py 
# Implements Int object
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################


###########################################################
#  class  Int
###########################################################   
class Int():
    def __init__(self, n: 'int' = 0):
        # ONLY DATA STRUCTURE ALLOWED
        # self._positive
        # self._a
        # print("Remove this line and Write code ")
        self._positive = True if n >= 0 else False
        self._a = [int(d) for d in str(abs(n))]

    def __str__(self):
        sign = '+' if self._positive else '-'
        return f"{sign}{self._a}"

    def __len__(self):
        return len(self._a)

    def __getitem__(self, index):
        return self._a[index]

    def __setitem__(self, index, value):
        self._a[index] = value

    def int(self):
        sign = 1 if self._positive else -1
        return sign * int(''.join(str(d) for d in self._a))

    def __add__(self, other):
        if isinstance(other, Int):
            a = self._a[::-1]
            b = other._a[::-1]
            len_a = len(a)
            len_b = len(b)
            len_max = max(len_a, len_b)
            carry = 0
            result = []
            for i in range(len_max):
                digit_sum = carry
                if i < len_a:
                    digit_sum += a[i]
                if i < len_b:
                    digit_sum += b[i]
                carry = digit_sum // 10
                digit = digit_sum % 10
                result.append(digit)
            if carry > 0:
                result.append(carry)
            result.reverse()
            return Int(self.int() + other.int())
        else:
            raise TypeError("Operand not supported for +")

    def __sub__(self, other):
        if isinstance(other, Int):
            return Int(self.int() - other.int())
        else:
            raise TypeError("Operand not supported for -")

    def __lt__(self, other):
        if isinstance(other, Int):
            return self.int() < other.int()
        else:
            raise TypeError("Operand not supported for <")

    def __gt__(self, other):
        if isinstance(other, Int):
            return self.int() > other.int()
        else:
            raise TypeError("Operand not supported for >")

    def __eq__(self, other):
        if isinstance(other, Int):
            return self.int() == other.int()
        else:
            raise TypeError("Operand not supported for ==")

    def __ne__(self, other):
        if isinstance(other, Int):
            return self.int() != other.int()
        else:
            raise TypeError("Operand not supported for !=")
    #############################
    # WRITE All public functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    #############################

    ##############################################################
    # WRITE All private functtions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    ##############################################################
    
############################################################
# Inttest.py 
# Test Bench for Int
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
#  NOTHING CAN BE CHANGED IN THIS FILE
########################################################### 

############################################################
#  All imports here
###########################################################
import sys
from typing import List # For getting Python Version
#from Int import *

############################################################
#  class  Int test
###########################################################  


class Inttest():
    def __init__(self):

        pass

    def data_structure(self):
        print("------------Testing data_structure-----------------")
        n = 67
        a = Int(n)
        print(a)
        a = Int(100)
        print(a)
        a = Int(-100)
        print(a)
        n = -10098765456786542342222
        a = Int(n)
        print(a)
        ans = [1, 0, 0, 9, 8, 7, 6, 5, 4, 5, 6, 7, 8, 6, 5, 4, 2, 3, 4, 2, 2, 2, 2]
        d = len(a)
        d1 = len(ans)
        if (d != d1):
            print("Number of digits is ", d1, " Your answer is", d)
        assert (d == d1)
        print("--------------- data_structure Passed --------------- ")

    def access(self):
        print("------------Testing access -----------------")
        n = -10098765456786542342222
        a = Int(n)
        print(a)
        ans = [1, 0, 0, 9, 8, 7, 6, 5, 4, 5, 6, 7, 8, 6, 5, 4, 2, 3, 4, 2, 2, 2, 2]
        d = len(a)
        d1 = len(ans)
        if (d != d1):
            print("Number of digits is ", d1, " Your answer is", d)
        assert (d == d1)
        for i in range(d):
            e = a[i]
            e1 = ans[i]
            if (e != e1):
                print("a[", i, "] = ", e1, "But your answer is", e)
            assert (e == e1)
        a[0] = 5;
        print(a)
        n = a.int()
        n1 = -50098765456786542342222
        if (n != n1):
            print("Correct ans=", n1, "But your answer is", n)
        assert (n == n1)
        print("--------------- access Passed --------------- ")

    def arithmetic(self):
        print("------------Testing arithmetic-----------------")
        q = [0, 5, -91, 1086, 1235657786899879757575175157511571, -12356577868998797575789107815]
        s = len(q)
        print("---------------  Testing +  --------------- ")
        for i in range(s):
            m = q[i]
            a = Int(m)
            print("a = ", end="")
            print(a)
            n = q[s - 1 - i]
            b = Int(n)
            print("b = ", end="")
            print(b)

            mn = m + n
            c = a + b
            print("c = ", end="")
            print(c)
            vc = c.int()
            if (vc != mn):
                print("Expected value is: ", mn, "But your answer is", vc);
                assert (0)

        print("--------------- + Passed --------------- ")

        print("---------------  Testing -  --------------- ")
        for i in range(s):
            m = q[i]
            a = Int(m)
            print("a = ", end="")
            print(a)
            n = q[s - 1 - i]
            b = Int(n)
            print("b = ", end="")
            print(b)

            mn = m - n
            c = a - b
            print("c = ", end="")
            print(c)
            vc = c.int()
            if (vc != mn):
                print("Expected value is: ", mn, "But your answer is", vc)
                assert (0)

        print("--------------- - Passed --------------- ")

        print("--------------- arithmetic Passed --------------- ")

    def relational(self):
        print("-------------Testing arithmetic-----------------")
        q1 = [0, 3, 2, -3, -2, 5, 5, -5, -5, -91, 1086, 1235657786899879757575175157511571,
              -12356577868998797575789107815, -12356577868998797575789107815]
        q2 = [-1, 2, 3, 2, 3, 5, -5, 5, -5, 91, 1085, 1235657786899879757575175157511570,
              -12356577868998797575789207815, -12356577868998797575789107815]
        s = len(q1)
        print("---------------  Testing <  --------------- ")
        for i in range(s):
            m = q1[i]
            a = Int(m)
            print("a = ", end="")
            print(a)
            n = q2[i]
            b = Int(n)
            print("b = ", end="")
            print(b)
            mn = m < n
            c = (a < b)
            assert (c == 0 or c == 1)
            print("is a < b", c)
            if (c != mn):
                print("Expected value is: ", mn, "But your answer is", c)
                assert (0)
        print("--------------- < Passed --------------- ")

        print("---------------  Testing >  --------------- ")
        for i in range(s):
            m = q1[i]
            a = Int(m)
            print("a = ", end="")
            print(a)
            n = q2[i]
            b = Int(n)
            print("b = ", end="")
            print(b)
            mn = (m > n)
            c = (a > b)
            assert (c == 0 or c == 1)
            print("is a > b", c)
            if (c != mn):
                print("Expected value is: ", mn, "But your answer is", c)
                assert (0)
        print("--------------- < Passed --------------- ")

        print("---------------  Testing ==  --------------- ")
        for i in range(s):
            m = q1[i]
            a = Int(m)
            print("a = ", end="")
            print(a)
            n = q2[i]
            b = Int(n)
            print("b = ", end="")
            print(b)
            mn = (m == n)
            c = (a == b)
            assert (c == 0 or c == 1)
            print("is a == b", c)
            if (c != mn):
                print("Expected value is: ", mn, "But your answer is", c)
                assert (0)
        print("--------------- == Passed --------------- ")

        print("---------------  Testing !=  --------------- ")
        for i in range(s):
            m = q1[i]
            a = Int(m)
            print("a = ", end="")
            print(a)
            n = q2[i]
            b = Int(n)
            print("b = ", end="")
            print(b)
            mn = (m != n)
            c = (a != b)
            assert (c == 0 or c == 1)
            print("is a != b", c)
            if (c != mn):
                print("Expected value is: ", mn, "But your answer is", c)
                assert (0)
        print("--------------- != Passed --------------- ")

############################################################
# MAIN
########################################################### 
def main():
    print("Basic Int test starts")
    test = Inttest()
    test.data_structure()
    test.access()
    test.arithmetic()
    test.relational()


if __name__ == '__main__':
    main()
```
